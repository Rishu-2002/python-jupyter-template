### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y values as the square of x

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot in the first subplot
axs[0, 0].plot(x, y, label='y = x^2', color='b', linewidth=2)
axs[0, 0].set_title('Line Plot of y = x^2')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Bar plot in the second subplot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Plot of Categories')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axs[1, 0].annotate(f'({txt[0]}, {txt[1]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightcoral', 'lightgreen'])
axs[1, 1].set_title('Pie Chart of Programming Languages')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
